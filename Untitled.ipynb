```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Social_Network_Ads.csv')

X = data.iloc[:, 2:4].values

X.shape

y = data.iloc[:, -1].values
# extracting last column and string as numpy array
y

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
# keeping 20% data as testing data

X_train.shape

X_test.shape

scaler = StandardScaler()
# this library is used so that age and salary column are in similar range

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)

# now we will start applying knn on out data

# the 1st step is to find the value of k.

np.sqrt(X_train.shape[0])
k = 17

# creating object of this class
knn = KNeighborsClassifier(n_neighbors=k)

# NOW WE WILL TRAIN OUR MODULE
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

y_pred.shape

# now we will compare y_pred with y_test to check weather predictions made using knn algo we correct or not
accuracy_score(y_test, y_pred)

# we can also find out the wrong falues predicted by out algo
confusion_matrix(y_test, y_pred)


def predict_output():
    age = int(input("Enter your age : "))
    salary = int(input("Enter your salary : "))

    X_new = np.array([[age], [salary]]).reshape(1, 2)
    X_new = scaler.transform(X_new)

    if knn.predict(X_new)[0] == 0:
        return "WILL NOT PURCHASE THE PRODUCT"
    else:
        return "WILL PURCHASE THE PRODUCT"


predict_output()
```
